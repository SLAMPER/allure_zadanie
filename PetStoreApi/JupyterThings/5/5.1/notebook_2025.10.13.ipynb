```python
%%file helpers.py

def number_grade_to_alpha_grade(num: float):
    if num >= 67.0:
        return 'A+'
    elif num >= 66.0:
        return 'A-'
    elif num >= 64.0:
        return 'A'
    elif num >= 63.0:
        return 'B+'
    elif num >= 62.5:
        return 'B'
    elif num >= 57.0:
        return 'B-'
    elif num >= 55.0:
        return 'C+'
    elif num >= 52.0:
        return 'C'
    elif num >= 50.0:
        return 'C-'
    elif num >= 44.0:
        return 'D+'
    elif num >= 40.0:
        return 'D'
    elif num >= 35.0:
        return 'D-'
    else:
        return 'F'
```

```python
%%file fixtures.py
import pytest
import csv

FILE_TO_OPEN = 'grades.csv'

@pytest.fixture(scope="function")
def get_opened_test_csv_file():
    with open(FILE_TO_OPEN, 'r', encoding='utf-8') as f:
        yield f

def get_rows():
    with open(FILE_TO_OPEN, 'r', encoding='utf-8') as f:
        reader = csv.DictReader(f, delimiter=',')
        return [x for x in reader]
```

```python
%%file test_grade_csv.py
import csv
import helpers as h
import pytest
from fixtures import get_opened_test_csv_file, get_rows


@pytest.mark.parametrize('row',get_rows())
def test_grade_csv_mean_to_final_grade(row):
    counter = 0
    summ = 0
    for key, val in row.items():
        if key.startswith('Test') or key == 'Final':
            summ += float(val)
            counter += 1

    mean = summ / counter
    mean_as_grade =  h.number_grade_to_alpha_grade(mean)

    assert 'Grade' in row
    grade = row['Grade']

    print(row)
    print(mean)
    print(mean_as_grade)
    assert mean_as_grade == grade
```
